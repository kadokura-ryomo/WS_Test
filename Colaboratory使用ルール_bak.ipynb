## Googleドライブに保存されたノートの「保存」を行う
Googleドライブに保存されたノートに対し、下記はRSリモート環境では実行不可。 
> [ファイル]タブ ⇒ [保存]

更新したファイルを保存する場合、GitHub上で保存を行うこと。手順は下記の「代用手段」に示す。

**◆代用手段**
1. 「GitHubにコピーを保存」、もしくは「コピーをGitHubGistとして保存」を実施する。

**◆補足**

「GitHubにコピーを保存」と「コピーをGitHubGistとして保存」の違い

*   「GitHubにコピーを保存」では、GitHubのリポジトリ/ブランチを選択し、名前をつけてノートブックを保存する。<br>
リポジトリ/ブランチ内に同名ファイルがある場合、新規作成ではなく変更扱いとなる。

*   「GitHubGistとして保存」では、現在のノートブックで新たなGitHubGistを作成する。「GitHubGistとして保存」をクリックするたびに新たなGitHubGistのリポジトリが作成される。
> GitHubGistはコード断片の共有サービスで、ソースコード1ファイル単位で管理を行うことができます。（複数もOK）<br>
> 新たにGitHubGistを作成するとそのGitHubGistのリポジトリが生成されます。<br><br>
> Colaboratoryから作成したGitHubGistはsecretとなり（※）検索できません。URLがわかれば誰でもアクセスできますが、URLはGitHub側で生成したハッシュ値のため見つけづらいものとなっています。<br>
>（※初期設定ではsecretでしたが、設定等で変えられるのかもしれません。未確認です。）<br><br>
> GitHubGistの説明は下記ページを参考にしました。：<br>
> ・https://qiita.com/YumaInaura/items/d8ba7ce702d844f11ae0<br>
> ・https://qiita.com/hkusu/items/18cbe582abb9d3172019<br>

## 「GitHubにコピーを保存」を行う
◆手順
1. [ファイル]タブ⇒[GitHubにコピーを保存]をクリックする。

3. 保存するレポジトリとブランチを選択する。レポジトリが存在しない場合は、作成する。（リポジトリの作成については「GitHubのリポジトリを作成する」を参照のこと）

<font color= "red">※既にGitHubのリポジトリ/ブランチに同名ファイルが存在する場合、新規作成ではなく変更と扱われます。</font>

## GitHubのリポジトリを作成する
ここでは簡単な作成までの手順までを説明する。
細かい設定を行いたい場合は、各自で対応すること。

◆手順
1.   GitHubにログインする
2.   Your repositoriesを開く。
> Your repositoriesの開き方:
> 下記画像①→②の順でクリックする。image.png



3. [New]ボタンをクリックする。
> image.png

4. リポジトリ作成の設定を行う。
> ①リポジトリ名<br><br>
> ②リポジトリの説明<br><br>
> ③公開範囲：Publicにする<br>
> ※Publicでない場合、Colaboratoryとの連携が不可となります。Colaboratoryと連携する場合には、必ずPublicに設定してください。<br><br>
> ④Readmeファイルの追加<br>
> ※Readmeファイルの追加を行うと、mainブランチが自動で作成されます。Readmファイルの追加を行わない場合は、ブランチが作成されません。<br>
> ※ブランチがないリポジトリに対してはColaboratoryから「GitHubにコピーの保存」が行えないため、別途ブランチを作成する必要があります。<br><br>
image.png
